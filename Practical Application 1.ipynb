```python
#DATA REPLACEMANT AND TYPE CONVERTION

coupons_df['destination'] = coupons_df['destination'].str.replace('No Urgent Place','Free') 
#print(coupons_df['destination'].value_counts())    

coupons_df['age'] = coupons_df['age'].str.replace('below21','20').replace('50plus','51').astype('Int64') 


#coupons_df['bar'] =coupons_df['bar'].str.replace('less1','never')
coupons_df['bar'] =coupons_df['bar'].str.replace('less1','never').replace('never','0').replace('1~3','1').replace('4~8','4').replace('gt8','9').astype('Int64')                                     #Replace less1 values to never in Bar column data
#print(coupons_df['Bar'].value_counts())


#coupons_df['coffee_house'] =coupons_df['coffee_house'].str.replace('less1','never')
coupons_df['coffee_house'] =coupons_df['coffee_house'].str.replace('less1','never').replace('never','0').replace('1~3','1').replace('4~8','4').replace('gt8','9').astype('Int64')                       #Replace less1 values to never in CoffeeHouse column data
#print(coupons_df['CoffeeHouse'].value_counts())


#coupons_df['carry_away'] = coupons_df['carry_away'].str.replace('less1','never')
coupons_df['carry_away'] = coupons_df['carry_away'].str.replace('less1','never').replace('never','0').replace('1~3','1').replace('4~8','4').replace('gt8','9').astype('Int64')                          #Replace less1 values to never in CarryAway column data
#print(coupons_df['CarryAway'].value_counts())


#coupons_df['RestaurantLessThan20'] = coupons_df['RestaurantLessThan20'].str.replace('less1','never')
coupons_df['RestaurantLessThan20'] = coupons_df['RestaurantLessThan20'].str.replace('less1','never').replace('never','0').replace('1~3','1').replace('4~8','4').replace('gt8','9').astype('Int64')    #Replace less1 values to never in RestaurantLessThan20 column data
#print(coupons_df['RestaurantLessThan20'].value_counts())

#coupons_df['Restaurant20To50'] = coupons_df['Restaurant20To50'].str.replace('less1','never')
coupons_df['Restaurant20To50'] = coupons_df['Restaurant20To50'].str.replace('less1','never').replace('never','0').replace('1~3','1').replace('4~8','4').replace('gt8','9').astype('Int64')            #Replace less1 values to never in Restaurant20To50 column data
#print(coupons_df['Restaurant20To50'].value_counts())

coupons_df['passanger'] = coupons_df['passanger'].str.replace('Friend(s)','Friends').replace('Kid(s)','Kids')    #Replace Friend(s) values to Friends, Kid(s) to Kids in passanger column data
#print(coupons_df['passanger'].value_counts())

coupons_df['coupon'] = coupons_df['coupon'].str.replace('Restaurant(<20)','Cheap Restaurant').replace('Carry out & Take away','Take Away').replace('Restaurant(20-50)','Expensive Restaurant') 
#print(coupons_df['coupon'].value_counts())


#00000 - 12500             12.000
#12500 - 24999             24.000
#25000 - 37499             37.000
#37500 - 49999.            49.000
#50000 - 62499             62.000
#62500 - 74999.            74.000
#75000 - 87499.            87.000
#87500 - 99999             99.000
#100000 -More 100000       110.000

coupons_df['income_usd'] = coupons_df['income_usd'].str.replace('Less than $12500','12000').replace('$12500 - $24999','24000').replace('$25000 - $37499','37000').replace('$37500 - $49999','49000').replace('$50000 - $62499','62000').replace('$62500 - $74999','74000').replace('$75000 - $87499','87000').replace('$87500 - $99999','99000').replace('$100000 or More','110000').astype('Int64')
#print(coupons_df['income'].value_counts())
```
